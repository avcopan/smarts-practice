```python
from rdkit import Chem
from rdkit.Chem import Draw

# Define the molecule
molecule = Chem.MolFromSmiles('CC(=O)OC1=CC=CC=C1C(=O)O')

# Define functional groups using SMARTS patterns
functional_groups = {
    'Carbonyl': '[CX3](=O)',
    'Hydroxyl': '[OX2H]',
    # 'Carboxylic Acid': '[CX3](=O)[OX2H1]',
    # 'Ester': '[CX3](=O)[OX2][CX3]',
    # 'Benzene': 'c1ccccc1'
}

# Find and highlight functional groups in the molecule
highlight_atoms = []
for group_name, smarts in functional_groups.items():
    patt = Chem.MolFromSmarts(smarts)
    matches = molecule.GetSubstructMatches(patt)
    for match in matches:
        highlight_atoms.extend(match)

# Draw the molecule with highlighted functional groups
img = Draw.MolToImage(molecule, highlightAtoms=highlight_atoms)
img.show()
```
